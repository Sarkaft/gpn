```python
# obtained by running:
# python explore_tokenizers.py


def makeArray(text):
    return np.fromstring(text[1:-1].replace("\n", ""), sep=' ', dtype=int)


df = pd.read_csv("tokenizer_statistics.tsv", sep="\t", index_col=0)
df.token_counts = df.token_counts.apply(makeArray)
df
```

```python
#plt.figure(figsize=(10, 5))
sns.barplot(data=df, x="tokenizer", y="compression")
plt.xticks(rotation=90);
```

```python
sns.barplot(data=df, x="tokenizer", y="proportion_singleton_tokens")
plt.xticks(rotation=90);
```

```python
for tokenizer_path in df.tokenizer.values:
    plt.plot(df.loc[tokenizer_path, "token_counts"], label=tokenizer_path)
    plt.legend()
    plt.show()
```
